## Codificación de algunas variables categoricas

Cada oportunidad puede tener más de un item 
y en particular los items tienen dos variables categoricas que son el nombre y la familia.
Cuando agrupemos por oportunidad buscaremos valores representativos de todos los items asociados a esa oportunidad.
Generalmente aplicamos funciones numericas, motivo por el cual necesitaremos codificar estas dos variables categóricas como números.

```python
import category_encoders as ce
encoder_family=ce.BinaryEncoder(cols=['Product_Family'])
df_train_bin1=encoder_family.fit_transform(df_train['Product_Family'])
df_train_bin1=df_train_bin1.add_prefix('Bit_')

encoder_name=ce.BinaryEncoder(cols=['Product_Name'])
df_train_bin2=encoder_name.fit_transform(df_train['Product_Name'])
df_train_bin2=df_train_bin2.add_prefix('Bit_')

df_train_encodeado=pd.concat([df_train,df_train_bin1,df_train_bin2],axis=1)
df_train_encodeado.head()

#from sklearn.preprocessing import LabelBinarizer
#lb = LabelBinarizer()
#df_train_bin=lb.fit_transform(df_train['Product_Family'])
#pd.concat([df_train,pd.DataFrame(df_train_bin)],axis=1)
```

```python
df_test_bin1=encoder_family.fit_transform(df_test['Product_Family'])
df_test_bin1=df_test_bin1.add_prefix('Bit_')

df_test_bin2=encoder_name.fit_transform(df_test['Product_Name'])
df_test_bin2=df_test_bin2.add_prefix('Bit_')

df_test_encodeado=pd.concat([df_test,df_test_bin1,df_test_bin2],axis=1)
df_test_encodeado.head()
```
