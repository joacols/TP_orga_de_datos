```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.legend import Legend
import matplotlib.patches as mpatches
import seaborn as sns
#from statsmodels.graphics.mosaicplot import mosaic
import io
import requests

%matplotlib inline


#import xgboost as xgb
#from xgboost import XGBClassifier
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
from sklearn import tree
from sklearn.metrics import log_loss

pd.set_option('display.max_columns', None)

```

```python
url_train="https://raw.githubusercontent.com/joacols/TP_orga_de_datos/main/Train_Preprocesado_TP2_Datos_2020-2C-encodeado.csv"
s=requests.get(url_train).content
df_train=pd.read_csv(io.StringIO(s.decode('utf-8')))
df_train
```

```python
url_test="https://raw.githubusercontent.com/joacols/TP_orga_de_datos/main/Test_Preprocesado_TP2_Datos_2020-2C-encodeado.csv"
s=requests.get(url_test).content
df_test=pd.read_csv(io.StringIO(s.decode('utf-8')))
df_test
```

## Separamos la variable a predecir

```python
X = df_train.drop(['Stage','Sales_Contract_No'], axis=1)
```

```python
y = df_train['Stage']
```

## Creamos set de entrenamiento y test


```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = \
    train_test_split(X, y, test_size=0.2, random_state=123)
```

## Random Forest Classifier (n_estimators=50, max_features="log2")


### Entrenamiento

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
clfRF = RandomForestClassifier(n_estimators=50, max_features="log2")
```

```python
clfRF.fit(X_train, y_train)
```

```python
y_pred = clfRF.predict(X_test)
```

```python
accuracy_score(y_test, y_pred)
```

```python
y_pred_probs = clfRF.predict_proba(X_test)
```

```python
log_loss(y_test, y_pred_probs)
```

### Prediccion

```python
clfRF = clfRF.fit(X, y)
```

```python
y_pred_probs =clfRF.predict_proba(df_test.drop(['Sales_Contract_No','Stage'], axis=1))[:,1]
```

```python
#Creo un dataframe con el formato ( id , precio (El precio predecido ))
prediccion = df_test['Opportunity_ID'].to_frame()
prediccion.insert(1,'Target', y_pred_probs , True) 
#Exportamos la prediccion en formato csv sin labels
prediccion.to_csv(r'./RandomForest_n_estimators_50_max_features_log2_complete_train_set.csv',index=False , header=True)
```
